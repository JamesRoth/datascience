```python
from numpy import radians, sin
from numpy import linspace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame()
df['years']=range(0,31)
df['value_float']=1000*(1.035)**df.years
df['value_str']=["$"+str(round(x)) for x in df.value_float]
```

```python
from numpy import radians, sin
from numpy import linspace
import matplotlib.pyplot as plt
import pandas as pd

plt.plot(df.value_float,df.years)
plt.xlabel('Money in account')
plt.ylabel('Years')
plt.show()
```

# Calculating Interest Over the Years

This notebook shows us the money in an account, starting at 1000 dollars at year 0, growing at an interest rate of 3.5%.
The interest is calculated using the equation: $ d  = p_o \times {(1.035)^y} $
